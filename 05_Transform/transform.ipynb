```python
from datetime import datetime, timedelta

def decode_timestamp(opd_date, act_time):
    opd_datetime = datetime.strptime(opd_date, "%d%b%Y:%H:%M:%S")
    act_timedelta = timedelta(seconds=act_time)
    timestamp_col = opd_datetime + act_timedelta
    return timestamp_col

data_col["TIMESTAMP"] = data_col.apply(lambda row: decode_timestamp(row["OPD_DATE"], row["ACT_TIME"]), axis=1)
data_col
```
